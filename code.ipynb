```
import PyPDF2
import pdfplumber
import json
import re
from pathlib import Path

# First, let's try to extract text using pdfplumber which is often better for text extraction
def extract_text_from_pdf(pdf_path):
    text_content = ""
    try:
        with pdfplumber.open(pdf_path) as pdf:
            for page in pdf.pages:
                page_text = page.extract_text()
                if page_text:
                    text_content += page_text + "\n"
        return text_content
    except Exception as e:
        print(f"Error with pdfplumber: {e}")
        return None

# Try extracting with pdfplumber first
pdf_path = "/workspace/uploads/Resume.pdf"
extracted_text = extract_text_from_pdf(pdf_path)

if extracted_text:
    print("Successfully extracted text using pdfplumber:")
    print("="*50)
    print(extracted_text[:2000])  # Show first 2000 characters
    print("="*50)
    print(f"Total text length: {len(extracted_text)} characters")
else:
    # Fallback to PyPDF2 if pdfplumber fails
    try:
        with open(pdf_path, 'rb') as file:
            pdf_reader = PyPDF2.PdfReader(file)
            extracted_text = ""
            for page in pdf_reader.pages:
                extracted_text += page.extract_text() + "\n"
        print("Successfully extracted text using PyPDF2:")
        print("="*50)
        print(extracted_text[:2000])  # Show first 2000 characters
        print("="*50)
        print(f"Total text length: {len(extracted_text)} characters")
    except Exception as e:
        print(f"Error with PyPDF2: {e}")
```

```
# Comprehensive PDF processing installation and text extraction
from metagpt.tools.libs.terminal import Terminal
import os
import json

print("=== COMPREHENSIVE PDF TO TEXT CONVERSION ===")

# Install all necessary system packages
print("1. Installing system dependencies...")
await terminal.run('apt-get update')
await terminal.run('apt-get install -y poppler-utils tesseract-ocr tesseract-ocr-eng imagemagick ghostscript')

# Install Python packages
print("2. Installing Python packages...")
await terminal.run('pip install pymupdf pdfplumber pdf2image pytesseract pillow pdfminer.six')

# Method 1: Try pdftotext (most reliable for text-based PDFs)
print("\n3. Attempting pdftotext extraction...")
try:
    result = await terminal.run('pdftotext -layout "/workspace/uploads/Resume.pdf" "/workspace/resume_text.txt"')
    
    if os.path.exists('/workspace/resume_text.txt'):
        with open('/workspace/resume_text.txt', 'r', encoding='utf-8') as f:
            extracted_text = f.read()
        
        if extracted_text and len(extracted_text.strip()) > 100:
            print("✓ SUCCESS with pdftotext!")
            print(f"Extracted {len(extracted_text)} characters")
            print("First 800 characters:")
            print("="*60)
            print(extracted_text[:800])
            print("="*60)
        else:
            print("pdftotext returned minimal text - PDF might be image-based")
            extracted_text = None
    else:
        extracted_text = None
except Exception as e:
    print(f"pdftotext failed: {e}")
    extracted_text = None

# Method 2: Try OCR extraction if pdftotext failed
if not extracted_text:
    print("\n4. Attempting OCR extraction (for image-based PDFs)...")
    try:
        # Convert PDF to images
        await terminal.run('pdftoppm -jpeg -r 300 "/workspace/uploads/Resume.pdf" "/tmp/page"')
        
        # Check generated images
        result = await terminal.run('ls /tmp/page*.jpg 2>/dev/null || echo "No images found"')
        
        if "No images found" not in result:
            print("✓ PDF converted to images successfully")
            
            # Extract text using tesseract
            extracted_text = ""
            result = await terminal.run('ls /tmp/page*.jpg')
            image_files = [f.strip() for f in result.split('\n') if f.strip()]
            
            for i, img_file in enumerate(sorted(image_files)):
                print(f"Processing page {i+1}...")
                ocr_result = await terminal.run(f'tesseract "{img_file}" stdout -l eng --psm 6')
                
                if ocr_result.strip():
                    extracted_text += f"\n=== PAGE {i+1} ===\n"
                    extracted_text += ocr_result.strip()
                    extracted_text += "\n\n"
            
            if extracted_text and len(extracted_text.strip()) > 100:
                print("✓ SUCCESS with OCR extraction!")
                print(f"Extracted {len(extracted_text)} characters via OCR")
                print("First 800 characters:")
                print("="*60)
                print(extracted_text[:800])
                print("="*60)
                
                # Save OCR result
                with open('/workspace/resume_text.txt', 'w', encoding='utf-8') as f:
                    f.write(extracted_text)
            else:
                print("OCR extraction failed to produce meaningful text")
                extracted_text = None
        else:
            print("Failed to convert PDF to images")
            extracted_text = None
            
    except Exception as e:
        print(f"OCR extraction failed: {e}")
        extracted_text = None

# Method 3: PyMuPDF as final attempt
if not extracted_text:
    print("\n5. Attempting PyMuPDF extraction...")
    try:
        import fitz
        
        doc = fitz.open("/workspace/uploads/Resume.pdf")
        extracted_text = ""
        
        for page_num in range(doc.page_count):
            page = doc[page_num]
            page_text = page.get_text()
            if page_text.strip():
                extracted_text += f"\n=== PAGE {page_num + 1} ===\n"
                extracted_text += page_text
                extracted_text += "\n"
        
        doc.close()
        
        if extracted_text and len(extracted_text.strip()) > 50:
            print("✓ SUCCESS with PyMuPDF!")
            print(f"Extracted {len(extracted_text)} characters")
            
            with open('/workspace/resume_text.txt', 'w', encoding='utf-8') as f:
                f.write(extracted_text)
        else:
            extracted_text = None
            
    except Exception as e:
        print(f"PyMuPDF extraction failed: {e}")
        extracted_text = None

# Final result
if extracted_text and len(extracted_text.strip()) > 50:
    print(f"\n✓ PDF TEXT EXTRACTION SUCCESSFUL!")
    print(f"✓ Text saved to /workspace/resume_text.txt")
    print(f"✓ Total characters extracted: {len(extracted_text)}")
    print(f"✓ Ready to parse resume information")
    
    # Show a preview of the extracted text
    print("\nFull extracted text preview:")
    print("="*80)
    print(extracted_text)
    print("="*80)
else:
    print("\n✗ All extraction methods failed")
    print("The PDF might be heavily encrypted, corrupted, or contain complex layouts")
```
